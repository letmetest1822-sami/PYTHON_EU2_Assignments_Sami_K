```python
''' Question001_BMI

*  BMI = mass / (height * height)
*  In this formula, mass is in kilograms and height is in meters. The health risk associated with a BMI value is
    - Underweight    <  18.5
    - Normal Weight  >= 18.5 and < 25
    - Overweight     >= 25   and < 30
    - Obese          >= 30 
    
'''

massPound=210;
heightFeet = 5;
heightInch=11; # Inputs are weight in pounds and height in feets and inches

massKg = massPound / 2.2;
heightMeter = (heightFeet * 12 * 0.0254) + (heightInch *  0.0254)

print("Your Height : {0:5.2f} m ".format(heightMeter))
print("Your Weight : {0:5.2f} Kgs.".format(massKg))
#Just in case to check if the conversion is OK more or less...

BMI = massKg/(heightMeter * heightMeter); #In this formula, mass is in kilograms and height is in meters.

if BMI < 18.5: 
        print("Your weight in pounds: {0:5.2f} ".format(massPound))
        print("Your height in feet followed by a space then additional inches: {} {} ".format(heightFeet, heightInch))
        print("Your BMI is {0:5.2f} ".format(BMI))
        print("Your risk factory is 'Underweight'")
        
elif (BMI >= 18.5) & (BMI < 25):
        print("Your weight in pounds: {0:5.2f}".format(massPound))
        print("Your height in feet followed by a space then additional inches: {} {} ".format(heightFeet, heightInch ))
        print("Your BMI is {0:5.2f} ".format(BMI) )
        print("Your risk factory is 'Normal Weight'")

elif (BMI >= 25) & (BMI < 30):
        print("Your weight in pounds: {0:5.2f} ".format(massPound))
        print("Your height in feet followed by a space then additional inches: {} {} ".format(heightFeet, heightInch ))
        print("Your BMI is {0:5.2f} ".format(BMI))
        print("Your risk factory is 'Overweight'")

elif BMI > 30:
        print("Your weight in pounds: {0:5.2f}".format (massPound))
        print("Your height in feet followed by a space then additional inches:{} {} ".format(heightFeet, heightInch ))
        print("Your BMI is {0:5.2f} ".format(BMI))
        print("Your risk factory is 'Obese'")
```

```python
""" Question003_leapYear

* Any year is input by the user. 
* Write a program to determine whether the year is a leap year or not.
* Leap Years are any year that can be evenly divided by 4. 
* A year that is evenly divisible by 100 is a leap year only if it is also evenly divisible by 400.

	Example : 
	1992      Leap Year
	2000      Leap Year
	1900      NOT a Leap Year

"""
year = 1900

if year%4 == 0   and    (year%100 == 0    and    year%400 == 0):
    print("Year  {} is a leap year.".format(year));
    
else :
    print("Year  {} is not a leap year.".format(year));
```
